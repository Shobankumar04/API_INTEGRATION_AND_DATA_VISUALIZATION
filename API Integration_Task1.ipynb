```python
import requests
```

```python
url = "https://jsonplaceholder.typicode.com/posts"
```

```python
response = requests.get(url)
response
```

```python
if response.status_code == 200:
    data = response.json()  # Parse the JSON response
    print(data)  # Output the data
else:
    print(f"Error: {response.status_code}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
titles = [post['title'] for post in data] 
post_ids = [post['id'] for post in data]
```

```python
df = pd.DataFrame({'Post ID': post_ids, 'Title': titles})
```

```python
plt.figure(figsize=(10, 6))
plt.barh(df['Post ID'], df['Title'].str[:30], color='skyblue')  # Display first 30 characters of titles
plt.xlabel('Post Titles')
plt.xticks(rotation=90,fontsize=6)
plt.ylabel('Post ID')
plt.title('Post Titles from JSONPlaceholder API')
plt.show()
```
